```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r'G:\25国赛\C题\附件.xlsx', sheet_name='男胎检测数据')


y_concentration = df['Y染色体浓度']

less_than_004 = (y_concentration < 0.04).sum()
greater_than_or_equal_004 = (y_concentration >= 0.04).sum()


total = len(y_concentration)
less_004_percentage = less_than_004 / total * 100
greater_004_percentage = greater_than_or_equal_004 / total * 100

plt.rcParams['figure.dpi'] = 300

plt.rcParams['font.sans-serif'] = ['SimHei']

sns.histplot(y_concentration, kde=True)

plt.axvline(x=0.04, color='r', linestyle='--')
plt.text(0.02, 50, f'<0.04: {less_004_percentage:.2f}%', rotation=90, color='r')
plt.text(0.06, 50, f'>=0.04: {greater_004_percentage:.2f}%', rotation=90, color='r')

plt.title('Y 染色体浓度分布密度图')
plt.xlabel('Y 染色体浓度')
plt.xticks(rotation=45)
plt.ylabel('频次')
plt.savefig(
    'Y染色体浓度分布密度图.png',
    bbox_inches='tight',
    dpi=300,
)
plt.show()
```
