```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def plot_scatter_and_calculate_proportion(x, y, pregnancy_boundary, i, chromosome_concentration_boundary=0.04):
    """
    绘制散点图并计算满足条件的点的比例

    参数:
    x (pd.Series): 检测孕期数据
    y (pd.Series): Y染色体浓度数据
    pregnancy_boundary (float): 检测孕期分界点
    chromosome_concentration_boundary (float): 染色体浓度分界点
    """
    # 设置图片清晰度
    plt.rcParams['figure.dpi'] = 300

    # 设置中文字体
    plt.rcParams['font.sans-serif'] = ['SimHei']

    # 找出检测孕期小于pregnancy_boundary且Y染色体浓度大于chromosome_concentration_boundary的点
    pregnancy_boundary = pregnancy_boundary * 7
    condition = (x < pregnancy_boundary) & (y > chromosome_concentration_boundary)

    # 绘制散点图，满足条件的点用红色，不满足条件的点用蓝色
    plt.scatter(x[~condition], y[~condition], c='blue', label='其他点')
    plt.scatter(x[condition], y[condition], c='red',
                label=f'检测孕期<{pregnancy_boundary}且Y染色体浓度>{chromosome_concentration_boundary}的点')

    # 添加分界线
    plt.axvline(x=pregnancy_boundary, color='green', linestyle='--', label='检测孕期分界线')
    plt.axhline(y=chromosome_concentration_boundary, color='orange',
                linestyle='--', label='Y染色体浓度分界线')

    # 添加图例、标题和坐标轴标签
    plt.legend()
    plt.xlabel('检测孕期')
    plt.xticks(rotation=45)
    plt.ylabel('Y染色体浓度')
    plt.title('检测孕期与Y染色体浓度散点图')

    # 计算满足条件的点的数量
    num_condition = condition.sum()

    # 计算总点数
    total_num = len(x)

    # 计算满足条件的点的比例
    proportion = num_condition / total_num

    # 输出比例
    print(f'满足条件的点的比例为: {proportion * 100:.2f}%')

    plt.savefig(
        '第{i}组'.format(i=i),
        dpi=300,
        bbox_inches='tight',
    )
    # 显示图表
    plt.show()
```

```python
df = pd.read_excel('工作簿4.xlsx')
df_1 = df[df['孕妇BMI'] > 20.703124]
df_1 = df_1[df_1['孕妇BMI'] < 29.46993863368847]

x1 = df_1['检测孕期']
y1 = df_1['Y染色体浓度']

plot_scatter_and_calculate_proportion(x1, y1, 16.8, 1)
```
